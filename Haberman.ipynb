```python
#haberman data is not label. so how to label the data?
columns=['patients_age','year_of_operation','no_of_axillary_nodes','classes']

haberman=pd.read_csv("haberman.csv", names=columns)

haberman.head()

```

```python
#(1.3) Pair plot 
# In which we can virulize the data in 2-D
```

```python
plt.close()
sns.set_style("whitegrid");
sns.pairplot(haberman,hue="classes",size=4);
plt.show()
```

```python
#(1.4) PDF and CDF
```

```python
sns.FacetGrid(haberman,hue="classes",size=5)\
    .map(sns.distplot,"no_of_axillary_nodes")\
    .add_legend()
plt.show();
```

```python
sns.FacetGrid(haberman,hue="classes",size=5)\
             .map(sns.distplot,"year_of_operation")\
             .add_legend() 
plt.show()
```

```python
sns.FacetGrid(haberman,hue="classes",size=5)\
              .map(sns.distplot,"patients_age")\
              .add_legend() 
plt.show()               
```

```python
five_more=haberman.loc[haberman["classes"]==1]
five_less=haberman.loc[haberman["classes"]==2]

counts,bin_edges=np.histogram(five_more['no_of_axillary_nodes'],bins=10,density=True)

pdf=counts/(sum(counts))

print(pdf);
print(bin_edges);

cdf=np.cumsum(pdf);

plt.plot(bin_edges[1:],pdf);
plt.plot(bin_edges[1:],cdf);

plt.show();
```

```python
counts,bin_edges=np.histogram(five_less[],bins=10,density=True)

pdf=counts/(sum(counts))

print(pdf);
print(bin_edges);

cdf=np.cumsum(pdf);

plt.plot(bin_edges[1:],pdf);
plt.plot(bin_edges[1:],cdf);

plt.show()
```
